```python
#importar librerias
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
CarData= pd.read_csv("./car data.csv")
```

```python
CarData.head()
```

```python
    #convertir datos en dataframe
    df = pd.DataFrame(CarData)
    x = df[['Selling_Price']] #variable idependiente
    y = df[['Present_Price']] #variable dependiente
    
    #crear el modelo de regresion lineal
    modelou = LinearRegression()
    
    #ajustar el modelo a los datos
    modelou.fit(x,y)
    
    #coeficientes de la regresion 
    
    beta_0 = modelou.intercept_
    beta_1 = modelou.coef_
    
    #mostrar los coeficientes
    
    print(f"intercepto b0: {beta_0}")
    print(f"coeficiente b1: {beta_1}")
    
    #visualizar la regresion
    
    plt.scatter(df['Selling_Price'],y, color='blue', label='Datos reales')
    plt.plot(df['Selling_Price'],modelou.predict(x), color ='red',label ='Linea de regresion')
    plt.title ('Regresion lineal simple')
    plt.xlabel('Selling_Price')
    plt.ylabel('Present_Price')
    plt.legend()
    plt.show()
```

REGRESION LINEAL COMPUESTA

```python
CarDatavd= pd.read_csv("./Car details v3.csv")
```

```python
#convertir datos en dataframe
df = pd.DataFrame(CarDatavd)
x = df[['year','seats']] #variable idependiente
y = df['selling_price'] #variable dependiente
#crear el modelo de regresion lineal
modelo = LinearRegression()  
#ajustar el modelo a los datos
modelo.fit(x,y)
    
#coeficientes de la regresion 
    
beta_0 = modelo.intercept_
beta_1, beta_2 = modelo.coef_
    
    #mostrar los coeficientes
    
print(f"intercepto b0: {beta_0}")
print(f"coeficiente b1: {beta_1}")
print(f"coeficiente b1: {beta_2}")
    
    #visualizar la regresion
    
plt.scatter(df['year'],y, color='blue', label='Datos reales')
plt.plot(df['year'],modelo.predict(x), color ='red',label ='Linea de regresion')
plt.title ('Regresion lineal Multiple')
plt.xlabel('Aos')
plt.ylabel('Precio de venta')
plt.legend()
plt.show()
```

```python
#predecir precios actuales 

nuevo_X= np.array ([[2024, 4], [2018, 4]])
predicciones = modelo.predict(nuevo_X)

#imprimir predicciones

print("\nPrediccioness:")
for i, pred in enumerate(predicciones):
   print(f"prediccin {i+1} para modelo {nuevo_X[i][0]},{nuevo_X[i][0]}:{pred:.2f}")
```
